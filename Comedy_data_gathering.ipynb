```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
```

```python
urls=[]
mulaney=pd.read_csv('F:\\Datasets\\scrapsfromtheloft_Mulaney_urls.csv', encoding='latin1')
aliwong= pd.read_csv('F:\\Datasets\\scrapsfromtheloft_Ali Wong_urls.csv', encoding='latin1')
hannahgadsby= pd.read_csv('F:\\Datasets\\scrapsfromtheloft_Hannah Gadsby_urls.csv', encoding='latin1')
iliza= pd.read_csv('F:\\Datasets\\scrapsfromtheloft_Iliza_urls.csv', encoding='latin1')
trevornoah= pd.read_csv('F:\\Datasets\\scrapsfromtheloft_Trevor Noah_urls.csv', encoding='latin1')

mulaney.rename(columns={'fusion-image-wrapper href':'URL'}, inplace=True)
hannahgadsby.rename(columns={'fusion-image-wrapper href':'URL'}, inplace=True)
aliwong.rename(columns={'entry-title href':'URL'}, inplace=True)
trevornoah.rename(columns={'entry-title href':'URL'}, inplace=True)
iliza.rename(columns={'entry-title href':'URL'}, inplace=True)

frames=[mulaney, aliwong, hannahgadsby, iliza, trevornoah]
df=pd.concat(frames, ignore_index=True)
df.drop(columns=['fusion-single-line-meta', 'updated'], inplace=True)

#df['Comedian']=df['entry-title'].str.slice(0,df['entry-title'].str.find(':'))
df['pos']= (df['entry-title'].str.find(':'))
df['Comedian']=df.apply(lambda x: x['entry-title'][0:x['pos']], axis=1)

df['pos2']= (df['entry-title'].str.find('('))
df['Special']=df.apply(lambda x: x['entry-title'][x['pos']+1:x['pos2']-1], axis=1)

df.drop(columns=['entry-title', 'pos', 'pos2'], inplace=True)

df['Transcript']=None
df.tail()
```

```python
urls=[]
def populate_urls(list):
        for i in range(0,len(list)):
            urls.append(list[i])
            
        return urls
```

```python

populate_urls(mulaney['URL'].values.tolist())
populate_urls((aliwong['URL'].values.tolist()))
populate_urls(hannahgadsby['URL'].values.tolist())
populate_urls(iliza['URL'].values.tolist())
populate_urls(trevornoah['URL'].values.tolist())

urls
```

```python
count=0
for url in urls: 
    page=requests.get(url).text
    soup=BeautifulSoup(page,'lxml')
    text = [p.text for p in soup.find(class_="post-content").find_all('p')]
    df['Transcript'][count]=text
    #print(text)
    count=count+1
df    
```

```python
df['full_script']=None
for i in range(len(df)):
    joined=' '.join(df['Transcript'][i])
    df['full_script'][i]=joined
df.drop(columns=['Transcript'], inplace=True)    
df.head()
```

```python
df['Comedian']=df['Comedian'].str.title()
df['Special']=df['Special'].str.title()
#df['full_script']=df['full_script'].str.lower()
df
```
